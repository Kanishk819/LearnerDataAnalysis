```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_excel('/content/Coursera_MASKED_data.xlsx')
data.head()
```

```python
data.info()
```

```python
pd.isnull(data).sum()
```

```python
# Count the number of 'Yes' values in the 'Completed' column
completed_yes_count = data['Completed'].value_counts()['Yes']
completed_no_count = data['Completed'].value_counts()['No']

print("Number of Employees that completed a undertaken course : ", completed_yes_count)
print("Number of Employees that did not completed a undertaken course : ", completed_no_count)
```

```python
# Count the number of unique persons in the 'Name' column
unique_persons_count = data['Name'].nunique()
unique_division_count = data['Division'].nunique()
unique_department_count = data['Department'].nunique()

print("Number of unique persons in the 'Name' column:", unique_persons_count)
print("Number of unique Divisions : ", unique_division_count)
print("Number of unique Department : ", unique_department_count)
```

```python
data = data.dropna(subset=['Enrollment Time'])
data.reset_index(drop=True, inplace=True)
data.info()
data.head()
```

```python
data.describe()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
# data.groupby('Completed').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
# plt.gca().spines[['top', 'right',]].set_visible(False)

ax = data.groupby('Completed').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
for bars in ax.containers:
  ax.bar_label(bars)
```

**From this we can conclude that the number of incomplete courses is more than that of completed courses**

```python
from matplotlib import pyplot as plt
data.plot(kind='scatter', x='Duration(in hrs)', y='Learning Hours Spent', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)

x_data = data['Duration(in hrs)']
y_data = data['Learning Hours Spent']

# Calculate mean, max, and min
mean_x = np.mean(x_data)
mean_y = np.mean(y_data)
max_x = np.max(x_data)
max_y = np.max(y_data)
min_x = np.min(x_data)
min_y = np.min(y_data)

print("Mean Duration(in hrs):", mean_x)
print("Mean Learning Hours Spent:", mean_y)
print("Max Duration(in hrs):", max_x)
print("Max Learning Hours Spent:", max_y)
print("Min Duration(in hrs):", min_x)
print("Min Learning Hours Spent:", min_y)
```

**From this plot we can identify the Minimum , Maximum and Mean Duration and Learning Hours Spent**

```python
import seaborn as sns
ax = sns.barplot(data = data , x = 'Learning Hours Spent' , hue = 'Completed')
# ax.set_ylim(0, 40)

for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

**From this graph we can conclude that on an average the 5.47964 hours were spent on Learning by people who completed the course and 1.37468 hours were spent by people who did not complete the course**

```python
import matplotlib.pyplot as plt
completed_counts = data['Completed'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(completed_counts, labels=completed_counts.index, autopct='%1.1f%%', colors=[ 'red' , 'green'])
plt.title('Completion Status')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Filter data to include only participants who have completed the course
completed_data = data[data['Completed'] == 'Yes']

# Explore the skills learned by participants who completed the course
completed_skills = completed_data['Skills Learned'].str.split(', ').explode().value_counts()

# Plot the top 10 most frequent skills learned by completed participants
top_skills = completed_skills.head(10)
plt.figure(figsize=(10, 6))
top_skills.plot(kind='barh', color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Skill')
plt.title('Top 10 Most Frequent Skills Learned by Completed Participants')
plt.show()

# Identify the most pursued course among completed courses
most_pursued_course = completed_data['Course Name'].value_counts().idxmax()
print("Most Pursued Course Among Completed Courses:", most_pursued_course)
```
